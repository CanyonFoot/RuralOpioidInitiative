### 2003 data
- Methadone: 
    - MM: Methadone/LAAM maintenance
    - DM: Methadone/LAAM detoxification
- Buprenorphine (Prescribe physicians)
- Naltrexone (Extended release - injectable naltrexone)

```python
df2003 = pd.read_csv("2003/2003.csv")
df2003
```

```python
def find_key(row, key):
    if key in row['Keys']:
        return True
    return False
```

```python
df2003['Methadone_maintenance'] = df2003.apply(lambda row: find_key(row, 'MM'), axis=1)
```

```python
df2003.groupby(['Postal_code', 'Methadone_maintenance']).size()
#https://stackoverflow.com/questions/45752601/how-to-do-a-conditional-count-after-groupby-on-a-pandas-dataframe
```
